```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cols=["fLength","fWidth","fSize", "fConc","fConc1","fAsym", "fM3Long","fM3Trans","fAlpha","fDist","class"]
df=pd.read_csv("magic04.data",names=cols)
df.head()
df["class"]=(df["class"]=="g").astype(int)
for label in cols[:-1]:
   plt.hist(df[df["class"]==1][label],color="orange",alpha=0.7,density=True,label='gamma')
   plt.hist(df[df["class"]==0][label],color="blue",alpha=0.7,density=True,label='hadron')
   plt.title(label)
   plt.xlabel(label)
   plt.ylabel("Frequency")
   plt.legend()
   plt.show()




```
